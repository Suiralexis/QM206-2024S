```python
import matplotlib.pyplot as plt

# Lista de pa√≠ses de Am√©rica Central
central_america = [
    "Belize", "Costa Rica", "El Salvador", "Guatemala", "Honduras", "Nicaragua", "Panama"
]

# Filtrar datos de Am√©rica Central
central_america_df = df[df['Entity'].isin(central_america)]

# 1. Gr√°fico de Barras: Emisiones promedio por pa√≠s en Am√©rica Central
emissions_by_country = central_america_df.groupby('Entity')['Annual_CO2_emissions_per_capita'].mean().sort_values()

plt.figure(figsize=(10, 6))
emissions_by_country.plot(kind='bar', color='teal')
plt.title('Emisiones Promedio de CO2 Per C√°pita en Am√©rica Central')
plt.xlabel('Pa√≠s')
plt.ylabel('Emisiones Promedio (Toneladas)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 2. Gr√°fico de Dispersi√≥n: Emisiones a lo Largo del Tiempo en Am√©rica Central
plt.figure(figsize=(10, 6))
for country in central_america:
    subset = central_america_df[central_america_df['Entity'] == country]
    plt.scatter(subset['Year'], subset['Annual_CO2_emissions_per_capita'], alpha=0.6, label=country)

plt.title('Emisiones de CO2 Per C√°pita en Am√©rica Central a lo Largo del Tiempo')
plt.xlabel('A√±o')
plt.ylabel('Emisiones Per C√°pita (Toneladas)')
plt.legend(title='Pa√≠s')
plt.grid()
plt.show()

# 3. Gr√°fico de L√≠nea: Tendencia de Emisiones Totales en Am√©rica Central
emissions_by_year = central_america_df.groupby('Year')['Annual_CO2_emissions_per_capita'].sum()

plt.figure(figsize=(10, 6))
plt.plot(emissions_by_year.index, emissions_by_year.values, marker='o', linestyle='-', color='purple')
plt.title('Tendencia de Emisiones Totales de CO2 Per C√°pita en Am√©rica Central')
plt.xlabel('A√±o')
plt.ylabel('Emisiones Totales (Toneladas)')
plt.grid()
plt.show()
```

```python
# Versi√≥n 1 (Yo): An√°lisis del inventario de gas natural en Google Colab

import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Subir archivo localmente
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Cargar el archivo
df_gas_local = pd.read_csv(file_name)

# Convertir REF_DATE a formato datetime
df_gas_local['REF_DATE'] = pd.to_datetime(df_gas_local['REF_DATE'])

# Filtrar datos con unidad de medida en Gigajoules
df_gigajoules = df_gas_local[df_gas_local['Statistics'] == 'Gigajoules']

# Promedio mensual del inventario de apertura
monthly_opening_inventory = df_gigajoules[df_gigajoules['Storage'] == 'Opening inventory'].groupby('REF_DATE')['VALUE'].mean()

# Gr√°fico de l√≠nea: Promedio del inventario de apertura
plt.figure(figsize=(12, 6))
plt.plot(monthly_opening_inventory.index, monthly_opening_inventory.values, label='Inventario de Apertura', color='blue')
plt.title('Promedio del Inventario de Apertura en Gigajoules (2016-2024)')
plt.xlabel('Fecha')
plt.ylabel('Gigajoules')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()

# Cambios acumulativos del inventario por a√±o
yearly_inventory_change = df_gigajoules[df_gigajoules['Storage'] == 'Inventory change'].groupby(df_gigajoules['REF_DATE'].dt.year)['VALUE'].sum()

# Gr√°fico de barras: Cambios acumulativos del inventario
plt.figure(figsize=(10, 6))
yearly_inventory_change.plot(kind='bar', color='orange', alpha=0.7)
plt.title('Cambio Acumulativo del Inventario por A√±o (Gigajoules)')
plt.xlabel('A√±o')
plt.ylabel('Cambio Acumulado (Gigajoules)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Inventario de cierre promedio por a√±o
closing_inventory_by_year = df_gigajoules[df_gigajoules['Storage'] == 'Closing inventory'].groupby(df_gigajoules['REF_DATE'].dt.year)['VALUE'].mean()

# Gr√°fico de l√≠nea: Inventario de cierre promedio por a√±o
plt.figure(figsize=(12, 6))
plt.plot(closing_inventory_by_year.index, closing_inventory_by_year.values, marker='o', linestyle='--', color='green', label='Inventario de Cierre')
plt.title('Inventario de Cierre Promedio por A√±o (Gigajoules)')
plt.xlabel('A√±o')
plt.ylabel('Gigajoules')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Subir el archivo localmente
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Cargar el archivo
df = pd.read_csv(file_name)

# Relaci√≥n entre el tama√±o del motor y las emisiones de CO2
plt.figure(figsize=(10, 6))
plt.scatter(df['Engine Size(L)'], df['CO2 Emissions(g/km)'], alpha=0.7, color='blue')
plt.title('Relaci√≥n entre el Tama√±o del Motor y las Emisiones de CO2')
plt.xlabel('Tama√±o del Motor (L)')
plt.ylabel('Emisiones de CO2 (g/km)')
plt.grid()
plt.tight_layout()
plt.show()

# Emisiones promedio por clase de veh√≠culo
average_emissions_vehicle_class = df.groupby('Vehicle Class')['CO2 Emissions(g/km)'].mean().sort_values()

plt.figure(figsize=(12, 6))
average_emissions_vehicle_class.plot(kind='bar', color='green', alpha=0.8)
plt.title('Emisiones Promedio de CO2 por Clase de Veh√≠culo')
plt.xlabel('Clase de Veh√≠culo')
plt.ylabel('Promedio de Emisiones (g/km)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Impacto del tipo de combustible en las emisiones de CO2
average_emissions_fuel_type = df.groupby('Fuel Type')['CO2 Emissions(g/km)'].mean().sort_values()

plt.figure(figsize=(10, 6))
average_emissions_fuel_type.plot(kind='bar', color='orange', alpha=0.8)
plt.title('Impacto del Tipo de Combustible en las Emisiones de CO2')
plt.xlabel('Tipo de Combustible')
plt.ylabel('Promedio de Emisiones (g/km)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name)

exceed_mcl = df[df['Result'] > df['MCL']].groupby('School')['Result'].count()

plt.figure(figsize=(12, 6))
exceed_mcl.plot(kind='bar', color='red', alpha=0.8)
plt.title('Cantidad de Muestras que Exceden el Nivel M√°ximo de Contaminante (MCL)')
plt.xlabel('Escuela')
plt.ylabel('N√∫mero de Muestras')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(df['Result'], bins=20, color='blue', alpha=0.7)
plt.title('Distribuci√≥n de los Niveles de Plomo en las Muestras')
plt.xlabel('Concentraci√≥n de Plomo (mg/L)')
plt.ylabel('Frecuencia')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

consumption_points = df['Consumption Point'].value_counts()

plt.figure(figsize=(8, 6))
consumption_points.plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='Pastel1')
plt.title('Proporci√≥n de Muestras por Punto de Consumo')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name)

exceed_mcl = df[df['Result'] > df['MCL']].groupby('School')['Result'].count()

plt.figure(figsize=(12, 6))
exceed_mcl.plot(kind='bar', color='red', alpha=0.8)
plt.title('Cantidad de Muestras que Exceden el Nivel M√°ximo de Contaminante (MCL)')
plt.xlabel('Escuela')
plt.ylabel('N√∫mero de Muestras')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(df['Result'], bins=50, range=(0, 0.5), color='blue', alpha=0.7)
plt.title('Distribuci√≥n de los Niveles de Plomo en las Muestras (Rango Ajustado)')
plt.xlabel('Concentraci√≥n de Plomo (mg/L)')
plt.ylabel('Frecuencia')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

consumption_points = df['Consumption Point'].value_counts()

plt.figure(figsize=(8, 6))
consumption_points.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange'])
plt.title('Proporci√≥n de Muestras por Punto de Consumo')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
